```python
import matplotlib.pyplot as plt
import numpy as np
from Linear_Regression_1D_GD import Linear_Regression
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 12, 13, 17])
```

```python
algo = Linear_Regression()
algo.fit(x, y)
```

```python
algo.score(x, y)
```

```python
y_pred = algo.predict(x)
```

```python
plt.plot(x, y_pred)
plt.scatter(x, y)
plt.show()
```
